```python

# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data=pd.read_csv(mouse_drug_data_to_load)


clinical_trial=pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
df=mouse_data.merge(clinical_trial, on="Mouse ID")

df.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumorVolume=df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean()
# Convert to DataFrame
tumorVolume=pd.DataFrame(tumorVolume)

# Preview DataFrame
tumorVolume.head(100)
```

```python

# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumorVolumeError=df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem()
# Convert to DataFrame
tumorVolumeError=pd.DataFrame(tumorVolumeError)
# Preview DataFrame
tumorVolumeError.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumorVolumetemp=tumorVolume.unstack()
tumorVolumeFinal=tumorVolumetemp.transpose()
tumorVolumeErrortemp=tumorVolumeError.unstack()
tumorVolumeErrorFinal=tumorVolumeErrortemp.transpose()
#We added a temporary data frame as a saftey mechanism, so that you can run this cell as many times as you want without
#and still get the same output. If you don't use a temporary transition variable, if you run it more than once, the dataframe
#Can get misformated
# Preview that Reformatting worked
```

```python
tumorVolumeFinal
```

```python
tumorVolumeErrorFinal
```

```python

# Generate the Plot (with Error Bars)
a=tumorVolumeFinal.index.values.tolist()
index=[]
for i in range(len(a)):
    index.append(a[i][1])

plt.figure(figsize=(10,10))
plt.grid()
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")

plt.scatter(index,tumorVolumeFinal["Capomulin"], marker='.')
plt.errorbar(index,tumorVolumeFinal["Capomulin"],tumorVolumeErrorFinal["Capomulin"])

#plt.scatter(index,tumorVolumeFinal["Ceftamin"], marker="v")
#plt.errorbar(index,tumorVolumeFinal["Ceftamin"],tumorVolumeErrorFinal["Ceftamin"])

plt.scatter(index,tumorVolumeFinal["Infubinol"], marker='1')
plt.errorbar(index,tumorVolumeFinal["Infubinol"],tumorVolumeErrorFinal["Infubinol"])

plt.scatter(index,tumorVolumeFinal["Ketapril"],marker='s')
plt.errorbar(index,tumorVolumeFinal["Ketapril"],tumorVolumeErrorFinal["Ketapril"])

#plt.scatter(index,tumorVolumeFinal["Naftisol"],marker='P')
#plt.errorbar(index,tumorVolumeFinal["Naftisol"],tumorVolumeErrorFinal["Naftisol"])

plt.scatter(index,tumorVolumeFinal["Placebo"],marker='+')
plt.errorbar(index,tumorVolumeFinal["Placebo"],tumorVolumeErrorFinal["Placebo"])

#plt.scatter(index,tumorVolumeFinal["Propriva"],marker='*')
#plt.errorbar(index,tumorVolumeFinal["Propriva"],tumorVolumeErrorFinal["Propriva"])

#plt.scatter(index,tumorVolumeFinal["Ramicane"],marker='x')
#plt.errorbar(index,tumorVolumeFinal["Ramicane"],tumorVolumeErrorFinal["Ramicane"])

#plt.scatter(index,tumorVolumeFinal["Stelasyn"],marker="D")
#plt.errorbar(index,tumorVolumeFinal["Stelasyn"],tumorVolumeErrorFinal["Stelasyn"])

#plt.scatter(index,tumorVolumeFinal["Zoniferol"],marker="<")
#plt.errorbar(index,tumorVolumeFinal["Zoniferol"],tumorVolumeErrorFinal["Zoniferol"])

plt.legend(['Capomulin','Infubinol','Ketapril','Placebo'], loc="upper left")


# Save the Figure
#Show the Figure

plt.savefig('Tumor Volume.png')
plt.show()
```

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
sites=df.groupby(["Drug","Timepoint"])["Metastatic Sites"].mean()
# Convert to DataFrame
sites=pd.DataFrame(sites)
# Preview DataFrame
sites.head(20)
```

```python
# Store the Standard Error of Metastatic Sites Grouped by Drug and Timepoint
sitesError=df.groupby(["Drug","Timepoint"])["Metastatic Sites"].sem()
# Convert to DataFrame
sitesError=pd.DataFrame(sitesError)
# Preview DataFrame
sitesError.head(20)
```

```python
# Minor Data Munging to Re-Format the Data Frames
sitesTemp=sites.unstack()

sitesFinal=sitesTemp.transpose()

sitesErrortemp=sitesError.unstack()
sitesErrorFinal=sitesErrortemp.transpose()
# Preview that Reformatting worked
sitesFinal
```

```python
# Calculate the percent changes for each drug
percentDF=tumorVolumeFinal.pct_change(periods=9)
# Display the data to confirm
percentDF

```

```python
# Store all Relevant Percent Changes into a Tuple
subset=percentDF[['Capomulin','Infubinol','Ketapril','Placebo']]
percentTuples=[tuple(x) for x in subset.values]
percentList=list(percentTuples[-1])
for i in range(len(percentList)):
    percentList[i]=percentList[i]*100
percentList
```
